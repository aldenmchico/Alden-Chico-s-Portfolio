```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('reference/Telco-Customer-Churn.csv')
data.info()
```

```python
data.head()
```

```python
# Handle the Datetime column
data['last_review'] = pd.to_datetime(data['last_review'])
data['last_review_month'] = [d.month for d in data['last_review']]
data['last_review_day'] = [d.day for d in data['last_review']]
data['last_review_year'] = [d.year for d in data['last_review']]
data = data.drop('last_review', axis=1)
data.head()
```

---

## Numerical Data

```python
# Extract all the numeric columns from the data
numerics = ['int16','int32','int64','float16','float32','float64']
numerical_vars = list(data.select_dtypes(include=numerics).columns)
data_num = data[numerical_vars]
data_num.head()
```

```python
# Impute 0 to missing numerical values
data_num = data_num.fillna(0)
```

```python
# Split the dataset into variables and target
X_num = data_num.drop('availability_365', axis=1)
y = data['availability_365']
```

```python
# Train-test split the dataset
from sklearn.model_selection import train_test_split

X_num_train, X_num_test, y_train, y_test = train_test_split(X_num, y, 
                                                            test_size=0.3, 
                                                            random_state=42)
X_num_train.shape, X_num_test.shape, y_train.shape, y_test.shape
```

```python
# Normalize and scale the data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_num_train_data = scaler.fit_transform(X_num_train)
X_num_train = pd.DataFrame(data=X_num_train_data, columns=X_num_train.columns)
```

```python
# Fit the training data to the Logistic Regression model
from sklearn.linear_model import Lasso, LogisticRegression
from sklearn.feature_selection import SelectFromModel

selector = SelectFromModel(LogisticRegression(C=1000, penalty='l1'))
_ = selector.fit(X_num_train, y_train)
```

```python
select_feat_bool = selector.get_support()
```

```python
selected_feat = X_num_train.columns[select_feat_bool]
selected_feat
```

```python
non_select_feat = [not b for b in select_feat_bool]
non_selected_feat = X_num_train.columns[non_select_feat]
non_selected_feat
```
